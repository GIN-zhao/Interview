```python
import pandas as pd
import matplotlib.pyplot as plt
# CSVDataFrame


tmp_file = 'ch4.txt'
tmp_file2 = 'co2.txt'

df = pd.read_table(tmp_file,delimiter='\s+',names=['site','year','month','value'])

df1 = pd.read_table(tmp_file2,delimiter='\s+',names=['site','year','month','value'])

# value = []
# for (index,column_data) in df.iterrows():
#     # value.append(column_data['value'])
#     print(column_data['value'])
# print(value)
# 
df['Date'] = pd.to_datetime(df[['year', 'month']].assign(day=1))
df1['Date'] = pd.to_datetime(df1[['year', 'month']].assign(day=1))

# 
plt.figure(figsize=(10, 6))
# ax1 = plt.subplots(2,2)
ax1 = plt.subplot(211)
ax1.plot(df['Date'][9:21], df['value'][9:21], marker='o',color='r',label='co2')
ax1.legend()
ax2 = plt.subplot(212)
ax2.plot(df1['Date'][18:30], df1['value'][18:30], marker='+',color='b',label='ch4')
ax2.legend()
plt.title('Value Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)  # x
plt.grid(True)
# plt.legend()
# plt.show()
plt.savefig('hw.svg',format='svg')
```
